## 1. Opérateurs de comparaison et Logique
Pour prendre des décisions, il faut comparer des choses.
- `==` (égal à), `!=` (différent de), `>` (supérieur), `<` (inférieur)
- `and` (et), `or` (ou), `not` (non)

```python
age = 18
print("Est majeur ?", age >= 18)
print("A un permis mais moins de 25 ans ?", age >= 18 and age < 25)
```

## 2. Les conditions : if, elif, else
Permettent d'exécuter du code uniquement si une condition est vraie. **Attention à l'indentation (les espaces au début de la ligne) !** C'est ainsi que Python délimite ses blocs de code.

```python
meteo = "pluie"

if meteo == "soleil":
    print("Je vais à la plage.")
elif meteo == "pluie":
    print("Je reste lire un livre.")
else:
    print("Je ne sais pas quoi faire.")
```

## 3. Les Boucles : while et for
- **while** : Répète tant qu'une condition est vraie.
- **for** : Parcourt les éléments d'une séquence (comme une liste).

```python
# Boucle for
fruits = ["pomme", "banane", "cerise"]
for fruit in fruits:
    print("J'aime les", fruit + "s")

# Boucle while
compteur = 10
while compteur > 0:
    print("Décompte :", compteur)
    compteur -= 1
print("Décollage !")
```

## 🏆 Défi du Jour 2 : Le Jeu du Juste Prix

1. Créez une variable `prix_secret` contenant le nombre 42.
2. Demandez à l'utilisateur de deviner le prix en utilisant `input("Devinez le prix : ")` (convertissez sa réponse en entier avec `int()`).
3. Tant que l'utilisateur n'a pas trouvé le bon prix :
   - Si son nombre est plus petit, affichez "C'est plus !".
   - Si son nombre est plus grand, affichez "C'est moins !".
   - Redemandez-lui un nombre.
4. S'il trouve, affichez "Bravo, vous avez gagné !"

```python
# Écrivez votre code pour le défi ici
prix_secret = 42
def demander_nombre():
    while True:
        saisie = input("Devinez le prix : ")
        try:
            return int(saisie)
        except ValueError:
            print("Vous devez saisir un nombre.")

user_input = demander_nombre()

while user_input != prix_secret:
    while user_input < 0:
        print("Oups ! Saisissez un nombre positif.")
        user_input = demander_nombre()

    if user_input < prix_secret:
        print("C'est plus !")
    elif user_input > prix_secret:
        print("C'est moins !")
    else:
        break
    user_input = demander_nombre()

print("Bravo, vous avez gagné !")
```
